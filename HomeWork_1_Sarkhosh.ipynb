### Ackley function
<br><br>

In mathematical optimization, the Ackley function is a non-convex function used as a performance test problem for optimization algorithms. It was proposed by David Ackley in his 1987 PhD Dissertation(From Wikipedia)

<br><br>

On a 2-dimensional domain it is defined by:

![title](img/al_.png)


Its global optimum point is:

![title](img/al2_.png)

```python
import numpy as np
from matplotlib import pyplot as plt
import math
from prettytable import PrettyTable


def ackley(param_1=None, param_2=None, number_of_point=1000, lower_limit=-3, upper_limit=3):
    if type(param_1) is not list and param_1 is not None:
        print("param1: {}\nparam2:{}\nackley result: {}".format(param_1, param_2, math.exp(1) + 20 - 20 * math.exp(
            -0.2 * math.sqrt(0.5 * (param_1 * param_1 + param_2 * param_2))) - math.exp(
            0.5 * (math.cos(2 * math.pi * param_1) + math.cos(2 * math.pi * param_2)))))
    else:
        param_1 = [np.random.uniform(lower_limit, upper_limit) for _ in range(number_of_point)]
        param_2 = [np.random.uniform(lower_limit, upper_limit) for _ in range(number_of_point)]
        ackley = np.empty([len([param_1, param_2][0])])
        for i in range(len([param_1, param_2][0])):
            ackley[i] = math.exp(1) + 20 - 20 * math.exp(-0.2 * math.sqrt(
                0.5 * ([param_1, param_2][0][i] * [param_1, param_2][0][i] + [param_1, param_2][1][i] *
                       [param_1, param_2][1][
                           i]))) - math.exp(
                0.5 * (math.cos(2 * math.pi * [param_1, param_2][0][i]) + math.cos(
                    2 * math.pi * [param_1, param_2][1][i])))
        t = PrettyTable(['param 1', 'param 2', 'ackley'])
        [t.add_row([param_1[index], param_2[index], ackley[index]]) for index in range(len(param_1))]
        print(t)
        fig = plt.figure()
        ax = fig.gca(projection='3d')
        ax.scatter(param_1, param_2, ackley, label='Ackley Function')
        plt.show()
```
